```python
def getBlocks(arr):
    start = 0
    end = -1
    result = []
    for i in range(0, len(arr)):
        if arr[i] == 1:
            end = i
            if start < end and end == (len(arr) - 1):
                result.append([start, end])
        else:
            if start < end:
                result.append([start, end])
            start = i + 1
    return result

sample = [1,1,1,0,0,0,1,1]
x = getBlocks(sample)
print(x)
```

```python
# import ch2 (accept chiller no, accept setpoint)
def rule_ch2(chiller_num, X, sp, thres):
    column_map = {
        1: 8,
        2: 30,
        3: 80
    }
    idx_CHWST = column_map[chiller_num]  # chilled water supply temperature
    CWST_setpoint = sp  # chilled water supply temperature setpoint

    # Constants
    th = thres
    # th = 0.5  # unit: degrees Celsius
    t_int = 15  # unit: minutes
    sz_X = X.shape
    t_row = 5  # each row represents a 5-min time interval
    t_int //= t_row  # no. of time intervals: 3

    # Warning: No. of rows < No. of time intervals
    if sz_X[1] < t_int:
        print("Warning rule_ch2(): X duration=", sz_X[0]*t_row,
              "min is shorter than the required t_int=", t_int*t_row, "min.")

    # Get boolean vector of rule violations
    # Formula: CWST + th < CWST_setpoint
    deltaT=CWST_setpoint-X[:,idx_CHWST]
    idxT = deltaT > th
    idx_violation = sw.find_idx(idxT)


    # For every set of violations, check the conditions of the rule
    y=np.zeros(shape=(sz_X[0],1), dtype=int)

    for n in range(len(idx_violation)):

        y[idx_violation[n]] = 1

    return y
```

```python
def rule_ch8(chiller_num, X, thres):
    column_map = {
        1: 16,
        2: 27,
        3: 74
    }
    
    idx_RUNSTAT = column_map[chiller_num]  # run status

    # Constants
    th = thres  # unit: ? should be 29+ < x < 40
    t_int = 10  # unit: minutes
    sz_X = X.shape
    t_row = 5  # each row represents a 5-min time interval
    t_int //= t_row  # no. of time intervals

    # Warning: No. of rows < No. of time intervals
    if sz_X[1] < t_int:
        print("Warning rule_ch8(): X duration=", sz_X[0]*t_row,
              "min is shorter than the required t_int=", t_int*t_row, "min.")
    # Get boolean vector of rule violations
    # Formula: run status > th
    idxT = np.nan_to_num(X[:,idx_RUNSTAT], copy=True) > th
    idx_violation = sw.find_idx(idxT)

    # For every violation set, check conditions of the rule
    y=np.zeros(shape=(sz_X[0],1), dtype=int)

    for n in range(len(idx_violation) - t_int):

        # Get 1 set of violations
        idx = idx_violation[n:n+t_int+1]

        # Perform checks of rule violations
        cond1 = np.all(idxT[idx])
        cond2 = sum(np.diff(idx)) == t_int

        # Flag datapoint if rule violated
        if cond1 and cond2:
            y[idx_violation[n+t_int]] = 1

    return y
```

```python
def rule_chwp1(chiller_num, X, thres, time=15):
    column_map = {
        1: [10, 19],
        2: [46,44],
        3: [89,76]
    }
    idx_DP = column_map[chiller_num][0]  # differential pressure
    idx_SP = column_map[chiller_num][1]  # setpoint

    # Constants
    th = thres  # unit: percent
    t_int = time  # unit: minutes
    sz_X = X.shape
    t_row = 5  # each row represents a 5-min time interval
    t_int //= t_row  # no. of time intervals

    # Warning: No. of rows < No. of time intervals
    if sz_X[0] < t_int:
        print("Warning rule_chwp1(): X duration=", sz_X[0]*t_row,
              "min is shorter than the required t_int=", t_int*t_row, "min.")

    # Get boolean vector of rule violations
    # Formula: DP < SP * (100-th)%
    idxT = X[:, idx_DP] < X[:, idx_SP] * (1 - th)
    idx_violation = sw.find_idx(idxT)

    # For every violation set, check conditions of the rule
    y=np.zeros(shape=(sz_X[0],1), dtype=int)

    for n in range(len(idx_violation) - t_int):

        # Get 1 set of violations
        idx = idx_violation[n:n+t_int+1]

        # Perform checks of rule violations
        cond1 = np.all(idxT[idx])
        cond2 = sum(np.diff(idx)) == t_int

        # Flag datapoint if rule violated
        if cond1 and cond2:
            y[idx_violation[n+t_int]] = 1

    return y
```

```python
def rule_chwp2(chiller_num, X, thres, time=15):
    column_map = {
        1: [10, 19],
        2: [46,44],
        3: [89,76]
    }
    idx_DP = column_map[chiller_num][0]  # differential pressure
    idx_SP = column_map[chiller_num][1]  # setpoint

    # Constants
    th = thres  # unit: percent
    t_int = time  # unit: minutes
    sz_X = X.shape
    t_row = 5  # each row represents a 5-min time interval
    t_int //= t_row  # no. of time intervals

    # Warning: No. of rows < No. of time intervals
    if sz_X[0] < t_int:
        print("Warning rule_chwp1(): X duration=", sz_X[0]*t_row,
              "min is shorter than the required t_int=", t_int*t_row, "min.")

    # Get boolean vector of rule violations
    # Formula: DP > SP * (100+th)%
    idxT = X[:, idx_DP] > X[:, idx_SP] * (1 + th)
    idx_violation = sw.find_idx(idxT)

    # For every violation set, check conditions of the rule
    y=np.zeros(shape=(sz_X[0],1), dtype=int)

    for n in range(len(idx_violation) - t_int):

        # Get 1 set of violations
        idx = idx_violation[n:n+t_int+1]

        # Perform checks of rule violations
        cond1 = np.all(idxT[idx])  # Condition 1
        cond2 = sum(np.diff(idx)) == t_int  # Condition 2

        # Flag datapoint if rule violated
        if cond1 and cond2:
            y[idx_violation[n+t_int]] = 1

    return y
```

```python
def rule_chwp6(chiller_num, X, thres, time=15):
    column_map = {
        1: 4,
        2: 25,
        3: 86
    }
    idx_VSD = column_map[chiller_num]  # variable speed drive

    # Constants
    th = thres  # unit: percent
    t_int = time  # unit: minutes
    sz_X = X.shape
    t_row = 5  # each row represents a 5-min time interval
    t_int //= t_row  # no. of time intervals

    # Warning: No. of rows < No. of time intervals
    if sz_X[0] < t_int:
        print("Warning rule_chwp6(): X duration=", sz_X[0]*t_row,
              "min is shorter than the required t_int=", t_int*t_row, "min.")

    # Get boolean vector of rule violations
    # Formula: VSD >= th(0.90)
    idxT = X[:, idx_VSD] >= th
    idx_violation = sw.find_idx(idxT)

    # For every violation set, check conditions of the rule
    y = np.zeros(shape=(sz_X[0],1), dtype=int)

    for n in range(len(idx_violation) - t_int):

        # Get 1 set of violations
        idx = idx_violation[n:n+t_int+1]

        # Perform checks of rule violations
        cond1 = np.all(idxT[idx])  # Condition 1
        cond2 = sum(np.diff(idx)) == t_int  # Condition 2

        # Flag datapoint if rule violated
        if cond1 and cond2:
            y[idx_violation[n+t_int]] = 1

    return y
```

```python
def rule_ct2(chiller_num, X, thres, time=15):
    column_map = {
        1: 12,
        2: 36,
        3: 82
    }
    
    idx_CWST = column_map[chiller_num]  # cooling water supply temperature else column 3
    idx_SP = 28.3
    th = thres

    # Constants
    t_int = time        # min
    sz_X = X.shape
    t_row = 5
    t_int //= t_row

    # Warn if no. of rows < no. of time intervals
    if sz_X[0] < t_int:
        print("Warning rule_chwp6(): X duration=", sz_X[0]*t_row, "min is shorter than the required t_int=", t_int*t_row, "min.")

    # Get list of indexes where VSD >= th(0.90)
    deltaT = idx_SP - X[:, idx_CWST]
    idxT = deltaT > th
    idx_violation = sw.find_idx(idxT)

    # For every violation, check for the 2 conditions indicated below.
    y=np.zeros(shape=(sz_X[0],1), dtype=int)

    for n in range(len(idx_violation) - t_int):

        # Get 1 set of violations
        idx = idx_violation[n:n+t_int+1]

        # Perform checks of rule conditions
        cond1 = np.all(idxT[idx])  # Condition 1
        cond2 = sum(np.diff(idx)) == t_int  # Condition 2

        # Flag datapoint if rule violated
        if cond1 and cond2:
            y[idx_violation[n+t_int]] = 1

    return y
```

```python
sample_ct2_label = rule_ct2(3, frame.values, 0.5)
ind = np.where(sample_ct2_label == 1)[0]
ind = pd.DataFrame({'Label': ind})
ind.to_csv('chiller3_ct2.csv', index=None, header=True)
```

```python
def rule_ct3(chiller_num, X, thres, time=15):
    column_map = {
        1: 12,
        2: 36,
        3: 82
    }
    
    idx_CWST = column_map[chiller_num]  # cooling water supply temperature else column 3
    idx_SP = 28.3 
    th = thres

    # Constants
    t_int = time        # min
    sz_X = X.shape
    t_row = 5
    t_int //= t_row

    # Warn if no. of rows < no. of time intervals
    if sz_X[0] < t_int:
        print("Warning rule_chwp6(): X duration=", sz_X[0]*t_row, "min is shorter than the required t_int=", t_int*t_row, "min.")

    # Get list of indexes where VSD >= th(0.90)
    deltaT = X[:, idx_CWST] - idx_SP 
    idxT = deltaT > th
    idx_violation = sw.find_idx(idxT)

    # For every violation, check for the 2 conditions indicated below.
    y=np.zeros(shape=(sz_X[0],1), dtype=int)

    for n in range(len(idx_violation) - t_int):

        # Get 1 set of violations
        idx = idx_violation[n:n+t_int+1]

        # Perform checks of rule conditions
        cond1 = np.all(idxT[idx])  # Condition 1
        cond2 = sum(np.diff(idx)) == t_int  # Condition 2

        # Flag datapoint if rule violated
        if cond1 and cond2:
            y[idx_violation[n+t_int]] = 1
    return y
```
